```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}, Balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self.balance:
            print("Insufficient balance")  
        else:
            self.balance -= amount
            print(f"Withdrew {amount}, Balance: {self.balance}")

    def get_balance(self):
        return self.balance

initial_balance = float(input("Enter initial balance: "))
account = BankAccount(initial_balance)

while True:
    print("\nChoose an operation:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")
    choice = input("Enter choice (1-4): ")

    if choice == "1":
        amt = float(input("Enter amount to deposit: "))
        account.deposit(amt)
    elif choice == "2":
        amt = float(input("Enter amount to withdraw: "))
        account.withdraw(amt)  # <-- prints "Insufficient balance" if withdrawal fails
    elif choice == "3":
        print(f"Current Balance: {account.get_balance()}")
    elif choice == "4":
        print("Exiting...")
        break
    else:
        print("Invalid choice, try again.")
```

```python
class Book:
    def __init__(self, title, author, rating):
        self.title = title
        self.author = author
        self.rating = rating

    def update_rating(self, new_rating):
        self.rating = new_rating

    def display_info(self):
        print(f"{self.title} by {self.author} - Rating: {self.rating}/5")

book1 = Book("Python Basics", "John Doe", 4.5)
book1.display_info()  
book2 = Book("Java Guide", "Jane Smith", 3.8)
book2.update_rating(4.2)
book2.display_info()  
```
